#PRIMER PRACTICA CON DATA SET "Factores de la Performance de los Estudiantes"

```python
from google.colab import drive
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

Se realiza la carga del data set.

```python
education = pd.read_csv("/content/drive/MyDrive/StudentPerformanceFactors.csv")
education
```

#Traemos el data set de drive, borramos las pocas lineas con datos vacios o nulos

```python
education1 = education.dropna()
education1.isna().sum()
education1
```

#Corrijiendo nombres de la columnas, sacandole los guiones.

```python
education1.columns = education1.columns.str.replace('_', ' ')
education1.head(10)
```

# Teniendo el data set limpio y ordenado, se comienza con el analisis y las respuestas a las preguntas formuladas.

```python
import seaborn as sns
# Boxplot de las calificaciones del examen
sns.boxplot(y='Exam Score', data=education1)

plt.title('Distribución de Calificaciones del Examen')
plt.ylabel('Calificación del Examen')
plt.xlabel('Estudiantes')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
horas_estudio = education1['Hours Studied']
calificaciones = education1['Exam Score']

scatter = plt.scatter(horas_estudio, calificaciones,
                      c=calificaciones,
                      cmap='viridis',
                      alpha=(calificaciones - min(calificaciones)) / (max(calificaciones) - min(calificaciones)),
                      s=20)

plt.colorbar(scatter, label='Calificación')

plt.title('Relación entre Horas de Estudio y Calificación')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación del Examen')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd


accesos_recursos = education1['Access to Resources']
calificacion_final = education1['Exam Score']

df_grouped = education1.groupby(['Access to Resources', 'Exam Score']).size().reset_index(name='counts')
plt.scatter(df_grouped['Access to Resources'], df_grouped['Exam Score'],
            s=df_grouped['counts'] * 20,
            alpha=0.5, color='green')

plt.title('Relación entre el acceso a recursos y la calificación final')
plt.xlabel('Acceso a Recursos')
plt.ylabel('Calificación Final')

plt.show()
```

```python
school_counts = education1['School Type'].value_counts()

plt.figure(figsize=(8, 5))
sns.barplot(x=school_counts.index, y=school_counts.values, palette='viridis')

plt.title('Cantidad de Estudiantes por Tipo de Escuela')
plt.xlabel('Tipo de Escuela')
plt.ylabel('Cantidad de Estudiantes')

plt.show()
```

```python
# Filtrar estudiantes de escuelas públicas
public_students = education1[education1['School Type'] == 'Public']

plt.figure(figsize=(8, 4))
sns.boxplot(x='Access to Resources', y='Exam Score', data=public_students, palette='viridis')

plt.title('Calificación Final según Acceso a Recursos para Estudiantes de Escuelas Públicas')
plt.xlabel('Acceso a Recursos')
plt.ylabel('Calificación Final')

plt.show()
```

```python
# Filtrar estudiantes de escuelas públicas
private_students = education1[education1['School Type'] == 'Private']

plt.figure(figsize=(8, 4))
sns.boxplot(x='Access to Resources', y='Exam Score', data=private_students, palette='viridis',
            order=['High', 'Medium', 'Low'])

plt.title('Calificación según Acceso a Recursos para Estudiantes de Escuelas Privadas')
plt.xlabel('Acceso a Recursos')
plt.ylabel('Calificación Final')

plt.show()
```
